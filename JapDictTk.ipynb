```python
import csv
import os
import tkinter as tk
from tkinter import messagebox
import random

def _savE():
    if os.path.isfile('JapaneseDict2.csv'): #判斷有沒有這個檔案
        File=open('JapaneseDict2.csv','r',encoding='utf-8-sig')
        myDict=dict(csv.reader(File))
    else:
        myDict={} #如果不存在，就給一個空的
    
    keY=enteR.get()
    valuE=texT.get("1.0",'end-1c')
    myDict[keY]=valuE
    
    File=open('JapaneseDict2.csv','w',newline="",encoding='utf-8-sig')
    writeR=csv.writer(File)
    for rowK,rowV in myDict.items():
         writeR.writerow([rowK,rowV])

    File.close()

    enteR.delete("0", tk.END)
    texT.delete("1.0", tk.END)

    
    
def _traN():
    if os.path.isfile('JapaneseDict2.csv'): 
        File=open('JapaneseDict2.csv','r',encoding='utf-8-sig')
        myDict=dict(csv.reader(File))
    else:
        myDict={} 
        
    keY=enteR.get()
    if keY in myDict:
        texT.delete("1.0", tk.END)
        texT.insert(tk.END, myDict.get(keY))
    else:
        texT.delete("1.0", tk.END)
        texT.insert(tk.END, '沒有這個日文字啦!!!')
    
        File.close()
        
def _deL():
    if os.path.isfile('JapaneseDict2.csv'): 
        File=open('JapaneseDict2.csv','r',encoding='utf-8-sig')
        myDict=dict(csv.reader(File))
    else:
        myDict={} 
        
    keY=enteR.get()
    if keY in myDict:
        del myDict[keY]

    else:
        texT.delete("1.0", tk.END)
        texT.insert(tk.END, '沒有這個日文字啦!!!')
    
    File=open('JapaneseDict2.csv','w',newline="",encoding='utf-8-sig')
    writeR=csv.writer(File)
    for rowK,rowV in myDict.items():
         writeR.writerow([rowK,rowV])

    File.close()

    enteR.delete("0", tk.END)
    texT.delete("1.0", tk.END)

def _shoW():
    if os.path.isfile('JapaneseDict2.csv'): 
        File=open('JapaneseDict2.csv','r',encoding='utf-8-sig')
        myDict=dict(csv.reader(File))
    else:
        myDict={}
        
        File.close()
           
    wiN1=tk.Tk()
    wiN1.title("詞庫")
    wiN1.geometry("600x400")
    
    sBar=tk.Scrollbar(wiN1)
    sBar.pack(side=tk.RIGHT,fill=tk.Y)

    texT1=tk.Text(wiN1,font=("Arial",20),height=40,yscrollcommand=sBar.set)
    texT1.pack(side=tk.BOTTOM,fill=tk.BOTH)
    sBar.config(command=texT1.yview)
    
    iTemS=myDict.items()
    for iTem in iTemS:
        texT1.insert(tk.END,iTem)
        texT1.insert(tk.END,"\n")
    
def _tesT():
    if os.path.isfile('JapaneseDict2.csv'): 
        with open('JapaneseDict2.csv', 'r', encoding='utf-8-sig') as file:
            myDict = dict(csv.reader(file))
    else:
        myDict = {}

    def _checK(jantesTword):
        taiwanworD = texT2.get("1.0", 'end-1c')

        correct_answer = myDict.get(jantesTword, '')
        if taiwanworD == correct_answer:
            texT2.insert(tk.END, "恭喜答對了!!!")
        else:
            texT2.delete("1.0", tk.END)
            texT2.insert(tk.END, f"答錯了!!! 正確答案是: {correct_answer}")

    def _conTI():
        jantesTword = random.choice(list(myDict))
        #print(jantesTword)
        enteR1.delete("0", tk.END)
        texT2.delete("1.0", tk.END)
        enteR1.insert(tk.END, jantesTword)

    wiN2 = tk.Toplevel(wiN)
    wiN2.title("測驗開始")
    wiN2.geometry("600x400")

    lbL2 = tk.Label(wiN2, text="日文:", fg="black", bg="white", font=("Arial", 16))
    lbL2.pack()

    enteR1 = tk.Entry(wiN2, font=("Arial", 20), width=20)
    enteR1.pack()

    lbL3 = tk.Label(wiN2, text="請輸入中文:", fg="black", bg="white", font=("Arial", 16))
    lbL3.pack()

    texT2 = tk.Text(wiN2, font=("Arial", 20), height=2, width=20)
    texT2.pack()

    btN7 = tk.Button(wiN2, text="確認", font=("Arial", 12), width=10, height=2, command=lambda: _checK(enteR1.get()))
    btN7.pack()

    btN8 = tk.Button(wiN2, text="繼續", font=("Arial", 12), width=10, height=2, command=_conTI)
    btN8.pack()

    jantesTword = random.choice(list(myDict))
    #print(jantesTword)
    enteR1.insert(tk.END, jantesTword)
       
def _closE():
    qQ=tk.messagebox.askokcancel('提示','確定要結束程式嗎?')
    if qQ:
        wiN.destroy() 
    
    
wiN=tk.Tk()
wiN.title("智閔的日文小字典")
wiN.geometry("800x600")

lbL1=tk.Label(wiN,text="日文:",fg="black",bg="white",font=("Arial",16))
lbL1.pack()

enteR=tk.Entry(wiN,font=("Arial",20), width=20)
enteR.pack()

lbL2=tk.Label(wiN,text="中文:",fg="black",bg="white",font=("Arial",16))
lbL2.pack()

texT=tk.Text(wiN,font=("Arial",20),height=2, width=20)
texT.pack()


btN1=tk.Button(wiN,text="儲存",font=("Arial",12),width=10,height=2,command=_savE)
btN1.pack()
btN2=tk.Button(wiN,text="翻譯",font=("Arial",12),width=10,height=2,command=_traN)
btN2.pack()
btN3=tk.Button(wiN,text="刪除",font=("Arial",12),width=10,height=2,command=_deL)
btN3.pack()
btN4=tk.Button(wiN,text="顯示全部",font=("Arial",12),width=10,height=2,command=_shoW)
btN4.pack()
btN5=tk.Button(wiN,text="測驗",font=("Arial",12),width=10,height=2,command=_tesT)
btN5.pack()
btN6=tk.Button(wiN,text="結束",font=("Arial",12),width=10,height=2,command=_closE)
btN6.pack()


wiN.mainloop()
```
